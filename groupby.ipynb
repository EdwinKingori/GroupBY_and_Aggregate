```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/AlexTheAnalyst/PandasYouTubeSeries/main/Flavors.csv")
```

```python
df
```

```python
groupby_frame = df.groupby('Base Flavor')
```

Running Popular Aggregate Functions to Analyze the Dataset

```python
groupby_frame.mean()
```

```python
df.groupby('Base Flavor').mean()
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').sum()
```

Grouping aggregate functions on multiple columns

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean', 'max','min','count','sum'], 
                               'Texture Rating': ['mean', 'max','min','count','sum']})
```

```python
df.groupby(['Base Flavor', 'Liked']).mean()
```

```python
df.groupby(['Base Flavor', 'Liked']).agg({'Flavor Rating': ['mean', 'max','min','count','sum']})
```
